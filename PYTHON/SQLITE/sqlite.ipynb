# SQL And SQLite
SQL (Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. 

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a Table
cursor.execute('''
Create Table if Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department Text
    )
''')

## Commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
## Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
               values('Abhishek',25,'Engineer')

''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Abhishek',25,'Engineer')

''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Abhishek',25,'Engineer')

''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Abhishek',25,'Engineer')

''')

## Commit the changes
connection.commit()
```

```python
## Querying the data from sqlite3 (from the table)
cursor.execute('Select * from employees')
rows = cursor.fetchall()

## print the queired data
for row in rows:
    print(row)
```

```python
## Update the data in the table
cursor.execute('''
UPDATE employees
Set age=35
where name='Kish'
''')

connection.commit()
```

```python
## Querying the data from sqlite3 (from the table)
cursor.execute('Select * from employees')
rows = cursor.fetchall()

## print the queired data
for row in rows:
    print(row)
```

```python
## Delete the data from the table
cursor.execute('''
Delete from employees
               where name='Bob'
''')

connection.commit()
```

```python
## Querying the data from sqlite3 (from the table)
cursor.execute('Select * from employees')
rows = cursor.fetchall()

## print the queired data
for row in rows:
    print(row)
```

```python
## close the connection
connection.close()
```
