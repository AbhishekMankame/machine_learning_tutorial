## Numpy
NumPy is a fundamental library for scientific computing in Python. It provides support for arrays and matrices, along with a collection of mathematical functions to operate on these data structures.

```python
import numpy as np

## create array using numpy
## create a 1D array
arr1 = np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
arr2 = np.array([1,2,3,4,5])
arr2.reshape((1,5))
```

```python
arr3=np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(arr3)
print(arr3.shape)
```

```python
arr3.reshape(1,10)
```

```python
arr3.reshape(10,1)
```

```python
arr3.reshape(5,2)
```

```python
np.arange(0,10,2)
```

```python
np.arange(0,10,11)
```

```python
np.ones((3,4))
```

```python
np.ones(3)
```

```python
## identity matrix
np.eye(3)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print("Array:\n",arr)
print("Shape:",arr.shape) # Output: (2,3)
print("Number of dimensions:",arr.ndim) # Output: 2
print("Size (number of elements):",arr.size) # Output: 6
print("Data type:",arr.dtype) # Output: int64 (may vary based on platform)
print("Item size (in bytes):",arr.itemsize) # Output: 8 (may vary based on platform)
```
