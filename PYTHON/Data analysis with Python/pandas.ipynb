## Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame.
A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data strucutre with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
## Pandas series is one-dimensional array-like object that can hold any data type. It is similar to a column in a table.
```

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
## Create a series from dictonary
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## create a DataFrame from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data frame from a List of Dictionaries
data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':32,'City':'Bangalore'},
    {'Name':'Alice','Age':32,'City':'Bangalore'},
    {'Name':'Bob','Age':32,'City':'Bangalore'}
    
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
