## Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words knows as a lemma. Stemming is important in natural language understanding(NLU) and natural langauge processing(NLP).

```python
## Classification Problem
## Comments of product is a positive review or negative review

## We will be having 'reviews' which is a text data
```

```python
words = ['eating','eats','eaten','writing','writes','programming','programs','history','finally','finalize']
```

```python
## PorterStemmer
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + '---->'+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
print(stemming.stem('sitting'))
```

## RegexpStemmer class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
reg_stemmer=RegexpStemmer('^ing|ing$|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('ingeating')
```

```python
reg_stemmer=RegexpStemmer('ing|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('going')
```

```python
reg_stemmer.stem('apple')
```
