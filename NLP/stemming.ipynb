## Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words knows as a lemma. Stemming is important in natural language understanding(NLU) and natural langauge processing(NLP).

```python
## Classification Problem
## Comments of product is a positive review or negative review

## We will be having 'reviews' which is a text data
```

```python
words = ['eating','eats','eaten','writing','writes','programming','programs','history','finally','finalize']
```

```python
## PorterStemmer
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + '---->'+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
print(stemming.stem('sitting'))
```
