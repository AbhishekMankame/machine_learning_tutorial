```python
paragraph = 'The old clock tower stood quietly at the edge of the town, its hands frozen at a time no one could remember. Every evening, as the sun dipped below the hills, a faint humming sound echoed through the empty streets, as if the tower was trying to speak. Children dared each other to get close, but no one stayed long enough to listen carefully. Some believed it was broken machinery, while others whispered that it carried stories from a forgotten past, waiting for someone patient enough to hear them.'
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
# downloading stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph) # Tokenization
```

```python
type(sentences)
```

```python
# Apply Stopwords And Filter And then Apply Stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the words into sentences
```
