### Building character level Language model - Given some text, predict the next character

```python
text = "hello world. this is a tiny language model."
import torch
```

```python
### Tokenization

# unique characters
chars = sorted(list(set(text)))
vocab_size = len(chars)

# mappings
stoi = {ch:i for i, ch in enumerate(chars)}
itos = {i:ch for ch, i in stoi.items()}

# encode & decode
def encode(s):
    return [stoi[c] for c in s]

def decode(nums):
    return ''.join([itos[n] for n in nums])

data = torch.tensor(encode(text))
```

```python
## Building Training Data 
X = data[:-1]
Y = data[1:]
```
