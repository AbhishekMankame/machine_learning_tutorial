### Building character level Language model - Given some text, predict the next character

```python
text = "hello world. this is a tiny language model."
import torch
```

```python
### Tokenization

# unique characters
chars = sorted(list(set(text)))
vocab_size = len(chars)

# mappings
stoi = {ch:i for i, ch in enumerate(chars)}
itos = {i:ch for ch, i in stoi.items()}

# encode & decode
def encode(s):
    return [stoi[c] for c in s]

def decode(nums):
    return ''.join([itos[n] for n in nums])

data = torch.tensor(encode(text))
```

```python
## Building Training Data 
X = data[:-1]
Y = data[1:]
```

```python
## Define the Model (This is the LLM)
import torch.nn as nn
import torch.nn.functional as F

class BirgamLM(nn.Module):
    def __init__(self,vocab_size):
        super().__init__()
        self.token_embedding=nn.Embedding(vocab_size, vocab_size)

    def forward(self,x):
        logits = self.token_embedding(x)
        return logits
```

```python
## Train the model
model = BirgamLM(vocab_size)
optimizer = torch.optim.AdamW(model.parameters(),lr=0.1)

for step in range(300):
    logits = model(X)
    loss = F.cross_entropy(logits,Y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if step%50 == 0:
        print(f"step {step}, loss {loss.item():.4f}")
```
